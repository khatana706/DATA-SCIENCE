```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
%matplotlib inline

from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('/content/income.csv')
```

```python
df.head()
```

```python
plt.scatter(df['Age'],df['Income($)'])
```

```python
km=KMeans(n_clusters=3)
km
y_predicted= km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
y_predicted= km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster']=y_predicted
```

```python
df.head()
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]

plt.scatter(df1.Age, df1['Income($)'], color='green')
plt.scatter(df2.Age, df2['Income($)'], color='red')
plt.scatter(df3.Age, df3['Income($)'], color='black')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
scaler=MinMaxScaler()
scaler.fit(df[['Income($)']])
df[['Income($)']] = scaler.transform(df[['Income($)']])
df
```

```python
scaler=MinMaxScaler()
scaler.fit(df[['Age']])
df[['Age']] = scaler.transform(df[['Age']])
df
```

```python
df
```

```python
km=KMeans(n_clusters=3)

y_predicted= km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
km.cluster_centers_
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]

plt.scatter(df1.Age, df1['Income($)'], color='green')
plt.scatter(df2.Age, df2['Income($)'], color='red')
plt.scatter(df3.Age, df3['Income($)'], color='black')

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*', label='blue')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
k_rng= range(1,10)
sse=[]

for k in k_rng:
  km = KMeans(n_clusters=k)
  km.fit(df[['Age','Income($)']])
  sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('k')
plt.ylabel('Sum of sequre error')

plt.plot(k_rng, sse)
```
