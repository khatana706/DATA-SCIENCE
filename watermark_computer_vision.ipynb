```python
img2 = cv2.imread("img2.jpg")
```

```python
img1 = cv2.resize(img1, (1000,1000))
```

```python
plt.imshow(img1)
```

```python
img2 = cv2.resize(img2,(600,600))
```

```python
plt.imshow(img2)
```

```python
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img2)
```

```python
img1.shape
```

```python
img2.shape
```

```python
x_offset = 1000-600
y_offset = 1000-600
```

```python
rows, cols, channels = img2.shape
```

```python
rows
```

```python
cols
```

```python
channels
```

```python
roi = img1[y_offset:1000, x_offset:1000]
```

```python
plt.imshow(roi)
```

```python
roi.shape
```

```python
img2.shape
```

```python
img2gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img2gray)
```

```python
img2gray.shape
```

```python
plt.imshow(img2gray, cmap='gray')
```

```python
mask_inv = cv2.bitwise_not(img2gray)
```

```python
mask_inv.shape
```

```python
plt.imshow(mask_inv, cmap='gray')
```

```python
import numpy as np
```

```python
white_bg = np.full(img2.shape, 255, dtype=np.uint8)
```

```python
bk = cv2.bitwise_or(white_bg, white_bg, mask = mask_inv)
```

```python
bk.shape
```

```python
plt.imshow(bk)
```

```python
plt.imshow(mask_inv, cmap='gray')
```

```python
 fg = cv2.bitwise_or(img2,img2, mask = mask_inv)
```

```python
plt.imshow(fg)
```

```python
fg.shape
```

```python
final_roi = cv2.bitwise_or(roi, fg)
```

```python
plt.imshow(final_roi)
```

```python
large = img1
small = final_roi
```

```python
large[y_offset:y_offset+small.shape[0],x_offset:x_offset+small.shape[1]] = small
```

```python
plt.imshow(large)
```
