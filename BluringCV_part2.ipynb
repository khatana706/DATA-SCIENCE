```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('bricks.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
def load_img():
  img = cv2.imread('bricks.jpg').astype(np.float32)/255
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  return img
```

```python
def display_img(img):
  fig = plt.figure(figsize=(15,10))
  ax = fig.add_subplot(111)
  ax.imshow(img)
```

```python
i = load_img()
```

```python
display_img(i)
```

```python
img = load_img()
gamma = 1/3
img2 = np.power(img, gamma)
display_img(img2)
```

```python
img = load_img()
cv2.putText(img, text = 'GURJAR', org=(20,60), fontFace=cv2.FONT_HERSHEY_COMPLEX, fontScale=1, color=(255,0,0), thickness=1) 
```

```python
display_img(img)
```

```python
kernal = np.ones((5,5), np.float32)/25
```

```python
kernal
```

```python
img2 = cv2.filter2D(img, -1, kernal)
```

```python
display_img(img2)
```

```python
img = load_img()
cv2.putText(img, text = 'GURJAR', org=(20,60), fontFace=cv2.FONT_HERSHEY_COMPLEX, fontScale=1, color=(255,0,0), thickness=1) 
display_img(img)
```

```python
blur_img = cv2.blur(img, ksize=(3,3))
display_img(blur_img)
```

```python
img = load_img()
cv2.putText(img, text = 'GURJAR', org=(20,60), fontFace=cv2.FONT_HERSHEY_COMPLEX, fontScale=1, color=(255,0,0), thickness=1) 
display_img(img)
```

```python
gaus_img = cv2.GaussianBlur(img, ksize=(5,5), sigmaX=10)
```

```python
display_img(gaus_img)
```
