#plot three color Histogram

```python
img = blue_bricks
color = ('b','g','r')
for i,col in enumerate(color):
  hist = cv2.calcHist([img],[i],None,[256],[0,256])
  plt.plot(hist, color=col) 
  plt.xlim([0,256])
plt.title('blue_bricks_image')
plt.show()  
```

```python
img = dark_horse
color = ('b','g','r')
for i,col in enumerate(color):
  hist = cv2.calcHist([img],[i],None,[256],[0,256])
  plt.plot(hist, color=col) 
  plt.xlim([0,30])
plt.title('dark_horse_image')
plt.show() 
```

```python
img = dark_rainbow
color = ('b','g','r')
for i,col in enumerate(color):
  hist = cv2.calcHist([img],[i],None,[256],[0,256])
  plt.plot(hist, color=col) 
  plt.xlim([0,256])
plt.title('dark_rainbow_image')
plt.show() 
```

#Masking


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('rainbow.jpg')
```

```python
plt.imshow(img)
```

```python
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img2)
```

```python
img.shape
```

```python
mask = np.zeros(img2.shape[:2],np.uint8)
```

```python
plt.imshow(mask, cmap='gray')
```

```python
mask[150:200, 50:300] = 255
```

```python
plt.imshow(mask, cmap='gray')
```

```python
masked_img = cv2.bitwise_and(img, img, mask=mask)
show_masked_img = cv2.bitwise_and(show_rainbow, show_rainbow, mask=mask)
```

```python
plt.imshow(show_masked_img)
```

```python
hist_mask_red = cv2.calcHist([img2], channels=[1], mask=mask, histSize=[256], ranges=[0,256])
```

```python
plt.plot(hist_mask_red)
```

```python
hist_mask_out = cv2.calcHist([img2], channels=[0], mask=None, histSize=[256], ranges=[0,256])
```

```python

plt.plot(hist_mask_out)
```
