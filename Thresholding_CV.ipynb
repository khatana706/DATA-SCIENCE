```python
def show_pic(img1):
  fig = plt.figure(figsize=(12,12))
  ax = fig.add_subplot(111)
  ax.imshow(img1, cmap='gray')
```

```python
show_pic(img1)
```

```python
ret, thresh1 = cv2.threshold(img, 127,255,cv2.THRESH_BINARY )
```

```python
show_pic(thresh1)
```

```python
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11,8)
```

```python
show_pic(th2)
```

```python
th3 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11,8)
```

```python
show_pic(th3)
```

```python
blend = cv2.addWeighted(thresh1, 0.7,th2, 0.3, 0)
```

```python
show_pic(blend)
```
